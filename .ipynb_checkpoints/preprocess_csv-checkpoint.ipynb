```python
transcripts_csv_3 = 'transcripts_csv_3'
concat_p = 'concat_p'

csv_num = 1

    
    
for csv in sorted(os.listdir(transcripts_csv_3)):

    if '.csv' in csv:

        os.chdir(transcripts_csv_3)

        read = pd.read_csv(csv)

        os.chdir('../')

                
        
    for csv_1 in sorted(concat_p):

        #os.chdir('concat_p')

        if '.csv' in csv_1:

            os.chdir(concat_p)

            read1 = pd.read_csv(csv_1)

            os.chdir('../')

            

            #os.chdir('../')

        df = pd.concat([read, read1], axis=1)
        


        os.chdir('concat_p_transcript')

        df.to_csv(f'p{csv_num}.csv', sep=',', index=False)


        csv_num += 2

        os.chdir('../')
        
        break

    
        
```

```python
csv_num = 1
for i in sorted(os.listdir('transcripts_csv_3')):

    if '.csv' in i:

        os.chdir('transcripts_csv_3')

        read = pd.read_csv(i)
        
        os.chdir('../')

        #print(read)


    for j in sorted(os.listdir('concat_p')):

        if '.csv' in j:

            os.chdir('concat_p')

            read1 = pd.read_csv(j)

            os.chdir('../')

            #print(read1)

            #break

        for col in read['reult/alternative/0/word/0/tartTime']:
            
            

            col1 = read1.iloc[:, 0]

            col2 = read1.iloc[:,23]

            while (col < col1):

                new_df = read1.iloc[:, 0:23]

                new2_df = read1.iloc[:, 23:]

                for i in read['results/channelTag']:

                    if i == 0:

                        pass

                    elif i < i.cummax():

                        df = pd.concat([read, new_df], axis = 1)

                    elif i == i.cummax():
                        df = pd.merge(df, new2_df, how = 'outer' )
                    else:
                        pass
                    
        os.chdir('concat_p_transcript')             
        df.to_csv(f'p{csv_num}.csv', sep=',', index=False)
        csv_num += 2
        
        os.chdir('../')



```
